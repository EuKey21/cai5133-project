```python
import pandas as pd

# Load the raw Yelp files
business_df = pd.read_json("yelp_dataset/yelp_academic_dataset_business.json", lines=True)
review_df = pd.read_json("yelp_dataset/yelp_academic_dataset_review.json", lines=True)

# Keep only businesses that are open
business_df = business_df[business_df["is_open"] == 1].copy()

# Keep only the columns needed for this project
business_df = business_df[
    ["business_id", "review_count", "stars", "categories"]
].copy()

review_df = review_df[
    ["review_id", "user_id", "business_id", "stars", "date", "text"]
].copy()

# Keep only reviews for businesses that are open
review_df = review_df[
    review_df["business_id"].isin(business_df["business_id"])
].copy()

# Convert review date from string to datetime
review_df["date"] = pd.to_datetime(review_df["date"])

# Save the filtered data to CSV
business_df.to_csv("processed_data/business_clean.csv", index=False)
review_df.to_csv("processed_data/review_clean.csv", index=False)

print("Saved filtered CSV files.")
print("Business shape:", business_df.shape)
print("Review shape:", review_df.shape)
```
