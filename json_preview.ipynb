# Yelp JSON Quick Preview
This notebook previews each JSON file in this folder by printing the headers (columns) and the first 5 rows.

```python
from pathlib import Path
import pandas as pd
from IPython.display import display


def count_jsonl_rows(path: str) -> int:
    """Count total rows in a JSONL file by counting newline characters."""
    total = 0
    with open(path, "rb") as f:
        for chunk in iter(lambda: f.read(1024 * 1024), b""):
            total += chunk.count(b"\n")
    return total


def _truncate_value(value: object, max_chars: int = 20) -> object:
    if pd.isna(value):
        return value
    text = str(value)
    if len(text) <= max_chars:
        return text
    return text[:max_chars] + "..."


def preview_json(path: str, nrows: int = 5, text_preview_chars: int = 20):
    print(f"\n=== {path} ===")
    total_rows = count_jsonl_rows(path)
    df = pd.read_json(path, lines=True, nrows=nrows)
    print(f"Total rows in file: {total_rows}")
    print(f"Rows loaded: {len(df)}")
    print(f"Headers ({len(df.columns)}):")
    print(list(df.columns))
    print(f"\nLoaded rows (Values truncated to first {text_preview_chars} chars):")

    preview_df = df.applymap(lambda v: _truncate_value(v, text_preview_chars))

    with pd.option_context(
        "display.max_rows", None,
        "display.max_columns", None,
        "display.max_colwidth", None,
        "display.width", 0,
    ):
        print(preview_df.to_string(index=False))


json_files = sorted([p.name for p in Path('.').glob('*.json')])
print('JSON files found:', json_files)
```
